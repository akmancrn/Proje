```python
def fix_cities(df):
    df.loc[df['city'] == 'Istanbul', 'city'] = 'istanbul'
    df.loc[df['city'] == 'Ankara', 'city'] = 'ankara'
    df.loc[df['city'] == 'Izmir', 'city'] = 'izmir'    
    df.loc[df['city'] == 'Bursa', 'city'] = 'bursa'
```

```python
df = pd.read_csv('city_traffic.csv')
df['city'] = df['city'].str.lower()
df['city'] = df['city'].str.strip()
fix_cities(df)
files = ['city_busdensity.csv','city_percip.csv','city_poulations.csv',
         'city_taxes.csv','city_walkscore.csv']

for file in files:
    df_csv = pd.read_csv(file)
    if 'city' in df_csv.columns:
        df_csv['city'] = df_csv['city'].str.strip() 
        df_csv['city'] = df_csv['city'].str.lower() 
        #Get state abreviation for these
    fix_cities(df)
    if 'city' in df_csv.columns:
        df = pd.merge(df, df_csv, on = ['city'], how = 'outer')
        print(file,'success')
    else:
        print(file, 'no state column')
df.to_csv('city_data.csv')
```

```python
corrs = {}
cols = list(df.iloc[:,1:].columns)
for x in cols:
    cols = [w for w in cols if w != x]
    for w in cols:
        corrs[(x, w)] = df[x].corr(df[w])
results = [(x, v.round(2)) for x, v in corrs.items() if corrs[x] > 0.5]
results = sorted(results, key=lambda x: x[1], reverse=True)
results
```

```python
interact = [x[0] for x in results[4:]]
interacts = []
for i in range(len(interact)):    
    col1 = interact[i][0]
    col2 = interact[i][1]
    col_interact = col1+'_x_'+col2
    interacts.append(col_interact)
    df[col_interact] = df[col1]*df[col2]
```
